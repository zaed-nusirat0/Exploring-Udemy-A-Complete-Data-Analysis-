```python
def bar_ploty_counts(data, column):
    column_count = (data[column].value_counts(normalize=True) * 100).sort_values(ascending=False)
    
    fig = px.bar(
        x=column_count.index, 
        y=column_count.values, 
        color=column_count.index,
        title=f'{column}',
        text=column_count.apply(lambda x: f"{x:.2f}%"),  
        labels={'x': f"{column}", 'y': 'Counts'},
        template='plotly_dark'
    )
    
    fig.update_traces(
        textposition="outside",  
        textfont={"size": 16, "family": "Consolas"}, 
        cliponaxis=False
    )
    
    fig.update_layout(
        xaxis=dict(
            tickmode='linear',
            dtick=1
        ),
        margin=dict(l=50, r=50, t=50, b=100), 
        title=dict(
            x=0.5,  
            font=dict(size=20)
        )
    )
    
    return fig.show()
```

```python
def bar_ploty_groupby(data, x,y,colors,title):    
    fig = px.bar(
        x=data[x], 
        y=data[y], 
        color=data[colors],
        title=f'{title}',
        text=data[y],  
        labels={'x': f"{x}", 'y': f"{y}"},
        template='plotly_dark'
    )
    
    fig.update_traces(
        textposition="outside",  
        textfont={"size": 16, "family": "Consolas"}, 
        cliponaxis=False
    )
    
    fig.update_layout(
        xaxis=dict(
            tickmode='linear',
            dtick=1
        ),
        margin=dict(l=50, r=50, t=50, b=100), 
        title=dict(
            x=0.5,  
            font=dict(size=20)
        )
    )
    
    return fig.show()
```
